## Creating an Word Embedding and using cosine similiarty

A simple way to get the required result for our task is to create embeddings and use cosine similiarity to get the result.

```python
# Extracting important details
amazon = amazon[['product_name','retail_price','discounted_price']]
flipkart = flipkart[['product_name','retail_price','discounted_price']]
```

I will be using `CountVectorizer` available in `scikit learn` package as Text preprocessing, tokenizing and filtering of stopwords are all included in CountVectorizer, which builds a dictionary of features and transforms documents to feature vectors.

- Since Products in amazon and flipkart have creating an embedding with this would be sufficient for our task.
- We can alternatively create BERT embeddings but it take much time and will need more processing power and will result in the same performance.

```python
# Importing cosine_similarity & CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
# Fitting in order to vectorize each product name.
cv.fit(pd.concat([amazon["product_name"], flipkart["product_name"]]))
```

```python
cosine_prod = cosine_similarity(cv.transform(amazon["product_name"]), cv.transform(flipkart["product_name"]))
```

```python
final_fk = []
final_az = []
for i in range(len(cosine_prod)):
    final_fk.append(flipkart.iloc[cosine_prod[i].argmax()])
    final_az.append(amazon.iloc[i])
```

```python
def res(final_res, company_names):
    result = []
    for final, company in zip(final_res, company_names):
        result.append(
            pd.DataFrame(final)
            .rename(columns={"product_name": f"Product name in {company}" , 
                             "retail_price" : f"retail price in {company}" ,
                             "discounted_price" : f"discounted price in {company}"})
            .reset_index(drop=True)
        )
    return pd.concat(result, axis=1)


res([final_fk, final_az], ["Flipkart", "Amazon"])
```

### Method 2 : Outer Join

Since all same products have same `uniq_id` we can exploit this information to our use and can simple do a outer join between two products and get the desired result.

```python
amazon.shape
```

```python
flipkart.shape
```

```python
# Outer Join columns on unique_id
joined = pd.merge(amazon,flipkart,on='uniq_id',how='outer')
```

```python
joined.shape
```

```python
joined.columns
```

```python
# Getting the required columns for our data
joined = joined[['product_name_x' , 'retail_price_x' , 'discounted_price_x',  'product_name_y' , 'retail_price_y' ,'discounted_price_y']]
```

```python
# Renaming the columns
joined.rename(columns={'product_name_x':'Product name in Amazon','retail_price_x':'Retail Price in Amazon','discounted_price_x':'Discounted Price in Amazon' , 'product_name_y':'Product name in Flipkart','retail_price_y':'Retail Price in Flipkart','discounted_price_y':'Discounted Price in Flipkart'}, inplace=True)
```
